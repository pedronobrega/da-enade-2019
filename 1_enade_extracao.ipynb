# Extração de Dados do Enade

```python
# https://download.inep.gov.br/microdados/Enade_Microdados/microdados_enade_2019.zip
```

```python
# Importar bibliotecas necessárias
import pandas as pd
import numpy as np
import zipfile
import requests
from io import BytesIO
import os
```

```python
# Criar um diretório para armazenar o conteúdo do ENADE
os.makedirs("./enade2019", exist_ok=True)
```

```python
# Define a URL
url = "https://download.inep.gov.br/microdados/Enade_Microdados/microdados_enade_2019.zip"
```

```python
# Faz o download do content
filebytes = BytesIO(
    requests.get(url).content
)
```

```python
# Descompacta de zip para a pasta enade2019
myzip = zipfile.ZipFile(filebytes)
myzip.extractall("./enade2019")
```

## Verificando os dados baixados

```python
# Abre os dados do ENADE como pandas DataFrame
enade = pd.read_csv(
    "./enade2019/microdados_enade_2019/2019/3.DADOS/microdados_enade_2019.txt",
    sep = ";",
    decimal = ","
)
```
